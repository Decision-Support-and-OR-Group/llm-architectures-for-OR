```python
import pandas as pd
import os
```

```python
gpt5_base_input_dir = "../assets/raw_data/gpt_5_1_raw"
gpt5_output_dir = "../assets/data_processed/gpt_5_1"

gpt5mini_base_input_dir = "../assets/raw_data/gpt_5_mini_raw"
gpt5mini_output_dir = "../assets/data_processed/gpt_5_mini"
```

```python
gpt5_excel_path = os.path.join(gpt5_base_input_dir, "experiment_summary_merged.xlsx")
gpt5mini_excel_path = os.path.join(gpt5mini_base_input_dir, "experiment_summary_merged.xlsx")
```

```python
# Wczytanie danych gpt
df_gpt = pd.read_excel(gpt5_excel_path)
print(f"Wczytano {len(df_gpt)} wierszy danych dla GPT-5.1")
```

```python
# Wczytanie danych gpt_mini
df_mini = pd.read_excel(gpt5mini_excel_path)
print(f"Wczytano {len(df_mini)} wierszy danych dla GPT-5-Mini")
```

# --- PREPROCESSING ---

```python
# Drop unused columns
df_gpt.drop(columns=["total_cost_usd","execution_result", "final_code"], inplace=True)
df_mini.drop(columns=["total_cost_usd","execution_result", "final_code"], inplace=True)
```

```python
def round_numeric_values(x, ndigits=2):
    try:
        val = float(x)
        return round(val, ndigits)
    except (ValueError, TypeError):
        return x
```

```python
# Zaokrąglanie wartości funkcji celu
df_gpt["ground_truth_answer"] = df_gpt["ground_truth_answer"].apply(round_numeric_values)
df_gpt["agent_objective_value"] = df_gpt["agent_objective_value"].apply(round_numeric_values)
# zaokrąglanie czasu
df_gpt["latency_sec"] = df_gpt["latency_sec"].apply(round_numeric_values, ndigits=3)
print("Preprocessing zakończony dla GPT-5.1")
```

```python
# Zmiana nazwy kolumn
df_mini['model'] = 'GPT-5-Mini'
df_gpt['model'] = 'GPT-5.1'
```

```python
df_gpt.head()
```

```python
# Zaokrąglanie wartości funkcji celu
df_mini["ground_truth_answer"] = df_mini["ground_truth_answer"].apply(round_numeric_values)
df_mini["agent_objective_value"] = df_mini["agent_objective_value"].apply(round_numeric_values)
# zaokrąglanie czasu
df_mini["latency_sec"] = df_mini["latency_sec"].apply(round_numeric_values, ndigits=3)
print("Preprocessing zakończony dla GPT-5-Mini")
```

```python
print("Zapisywanie danych GPT-5.1 do pliku")
gpt5_output_excel_path = os.path.join(gpt5_output_dir, "experiment_summary.xlsx")
df_gpt.to_excel(gpt5_output_excel_path, index=False)
```

```python
print("Zapisywanie danych GPT-5-Mini do pliku")
gpt5mini_output_excel_path = os.path.join(gpt5mini_output_dir, "experiment_summary.xlsx")
df_mini.to_excel(gpt5mini_output_excel_path, index=False)
```
